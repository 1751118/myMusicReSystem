```python
import pandas as pd
import numpy as np

# 24036首歌，361346个LEs
train_data = pd.read_csv('./Train_Test_Splits/Context/train.csv')
test_data = pd.read_csv('./Train_Test_Splits/Context/test.csv')
df = pd.concat([train_data, test_data], axis=0)
df.info()
```

```python
spotify_df = pd.read_csv('./data/spotify_dataset.csv')

spotify_df.info()
```

```python
spo_drop = spotify_df.drop_duplicates(subset='track_id')
spo_drop.shape
```

```python
spo_drop.drop('Unnamed: 0', axis = 1, inplace = True)
spo_drop.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
df['track_id'] = lb.fit_transform(df['track_id'])
df_indices = df['track_id'].value_counts()
df_indices
```

```python
# 根据数量信息对 spotify_df 进行采样
sampled_df = spo_drop.iloc[np.repeat(df_indices.index, df_indices.values)].copy()
sampled_df['track_id'].value_counts()
```

```python
df['track_id'] = lb.inverse_transform(df['track_id'])
df['track_id'].value_counts()
```

```python
series_count_df = df['track_id'].value_counts()
series_count_spotify = sampled_df['track_id'].value_counts()
columns = ['track_id', 'instrumentalness', 'liveness', 'speechiness', 'danceability', 'valence', 'loudness', 'tempo', 'acousticness', 'energy', 'mode', 'key']
```

```python
# 使用 loc 方法替换 df 中符合条件的行
# df.loc[df['track_id'] == 'acca7456f99b4699871d3fe249e5fe11', ['track_id', 'tempo', 'liveness']] = df_new.iloc[0]
df.reset_index(drop=True, inplace=True)
sampled_df.reset_index(drop=True, inplace=True)
df, sampled_df
```

```python
mapping = dict(zip(series_count_df.index, series_count_spotify.index))
mapping
```

```python
df['track_id'] = df['track_id'].map(mapping)
df_merge = df.merge(spo_drop, on='track_id', how='left')
df_merge
```

```python
df_merge.info()
```

```python
# 假设df是你的DataFrame
filtered_df = df_merge[[col for col in df_merge.columns if not col.endswith('_x')]]
filtered_df.info()
```

```python
# 假设df是你的DataFrame
filtered_df.columns = [col[:-2] if col.endswith('_y') else col for col in filtered_df.columns]
filtered_df.info()
```

```python
filtered_df = filtered_df[[col for col in filtered_df.columns if col in df.columns]]
filtered_df.info()
```
